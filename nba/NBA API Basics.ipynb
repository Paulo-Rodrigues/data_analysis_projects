```python
from nba_api.stats.static import teams
nba_teams = teams.get_teams()
print(f'Number of teams {len(nba_teams)}')
```

```python
from nba_api.stats.static import players
nba_players = players.get_players()
print(f'Number pf players {len(nba_players)}')
```

```python
lakers = [team for team in nba_teams if team['full_name'] == 'Los Angeles Lakers']
lakers
```

```python
def get_by_nickname(nickname):
    return [team for team in nba_teams if team['nickname'] == nickname]
```

```python
raptors = get_by_nickname('Raptors')
raptors
```

```python
def get_player_by_name(full_name):
    return [player for player in nba_players if player['full_name'] == full_name]
```

```python
lebron = get_player_by_name('LeBron James')
lebron
```
